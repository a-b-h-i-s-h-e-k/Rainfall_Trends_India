Now, using statistical thresholds (1.5 standard deviations below or above the mean), let’s identify years with extreme or deficient rainfall. This will help detect drought years and periods of excessive rainfall

```python
from scipy.stats import pearsonr  # Import Pearson correlation function for calculating correlations between variables

# Identifying drought and extreme rainfall years

# Calculate the mean annual rainfall across all years
mean_rainfall = rainfall_data['ANNUAL'].mean()  # Used to identify deviations from average rainfall

# Calculate the standard deviation of annual rainfall
std_dev_rainfall = rainfall_data['ANNUAL'].std()  # Measures the variability of rainfall from the mean

# Identify drought years where annual rainfall is significantly lower than average
# A drought year is defined as having rainfall below (mean - 1.5 * standard deviation)
drought_years = rainfall_data[
    rainfall_data['ANNUAL'] < (mean_rainfall - 1.5 * std_dev_rainfall)
]

# Identify extreme rainfall years where annual rainfall is significantly higher than average
# An extreme rainfall year is defined as having rainfall above (mean + 1.5 * standard deviation)
extreme_rainfall_years = rainfall_data[
    rainfall_data['ANNUAL'] > (mean_rainfall + 1.5 * std_dev_rainfall)
]

# Correlating seasonal rainfall with annual rainfall totals

# List of columns representing seasonal rainfall data
seasonal_columns = ['Jan-Feb', 'Mar-May', 'Jun-Sep', 'Oct-Dec']

# Calculate Pearson correlation coefficients between each season's rainfall and annual rainfall
# Pearson correlation measures the strength and direction of the linear relationship between two variables
seasonal_correlations = {
    season: pearsonr(rainfall_data[season], rainfall_data['ANNUAL'])[0]  # [0] retrieves the correlation coefficient
    for season in seasonal_columns
}

# Displaying results for drought/extreme years and correlations

# Create a summary of drought years with the corresponding year and annual rainfall
# Resetting the index ensures the table is neat and easy to read
drought_years_summary = drought_years[['YEAR', 'ANNUAL']].reset_index(drop=True)

# Create a summary of extreme rainfall years with the corresponding year and annual rainfall
# Similarly, the index is reset for cleaner presentation
extreme_rainfall_years_summary = extreme_rainfall_years[['YEAR', 'ANNUAL']].reset_index(drop=True)

# Convert the seasonal correlations dictionary into a DataFrame for easier interpretation
# The keys of the dictionary (seasons) become the index, and the correlation coefficients are in the 'Correlation' column
seasonal_correlations_summary = pd.DataFrame.from_dict(
    seasonal_correlations, orient='index', columns=['Correlation']
)

# Return the summaries as outputs to review
drought_years_summary, extreme_rainfall_years_summary, seasonal_correlations_summary
```

The analysis identifies five significant drought years (e.g., 2002 and 2009) and seven extreme rainfall years (e.g., 1917 and 1990) based on deviations from the mean annual rainfall. Seasonal rainfall correlations with annual totals reveal that the monsoon season (June-September) has the strongest correlation (0.93), which indicates it predominantly drives annual rainfall patterns. In contrast, other seasons like January-February (0.23) and March-May (0.31) have weaker correlations, which emphasizes the critical role of the monsoon in India’s overall rainfall dynamics.

# What Is Isolation Forest?
Isolation forest is a machine learning algorithm for anomaly detection. It’s an unsupervised learning algorithm that identifies anomaly by isolating outliers in the data.

Isolation Forest is based on the Decision Tree algorithm. It isolates the outliers by randomly selecting a feature from the given set of features and then randomly selecting a split value between the max and min values of that feature. This random partitioning of features will produce shorter paths in trees for the anomalous data points, thus distinguishing them from the rest of the data.

In general the first step to anomaly detection is to construct a profile of what’s “normal”, and then report anything that cannot be considered normal as anomalous. However, the isolation forest algorithm does not work on this principle; it does not first define “normal” behavior, and it does not calculate point-based distances.

As you might expect from the name, Isolation Forest instead works by isolating anomalies explicitly isolating anomalous points in the dataset.

The Isolation Forest algorithm is based on the principle that anomalies are observations that are few and different, which should make them easier to identify. Isolation Forest uses an ensemble of Isolation Trees for the given data points to isolate anomalies.

Isolation Forest recursively generates partitions on the dataset by randomly selecting a feature and then randomly selecting a split value for the feature. Presumably the anomalies need fewer random partitions to be isolated compared to “normal” points in the dataset, so the anomalies will be the points which have a smaller path length in the tree, path length being the number of edges traversed from the r# oot node.

# Define and Fit Model
We’ll create a model variable and instantiate the IsolationForest class. We are passing the values of four parameters to the Isolation Forest method, listed below.

Number of estimators: n_estimators refers to the number of base estimators or trees in the ensemble, i.e. the number of trees that will get built in the forest. This is an integer parameter and is optional. The default value is 100.

Max samples: max_samples is the number of samples to be drawn to train each base estimator. If max_samples is more than the number of samples provided, all samples will be used for all trees. The default value of max_samples is ‘auto’. If ‘auto’, then max_samples=min(256, n_samples)

Contamination: This is a parameter that the algorithm is quite sensitive to; it refers to the expected proportion of outliers in the data set. # Why Use contamination=0.05?
Assumption About Anomaly Proportion:

By setting contamination=0.05, we assume that 5% of the dataset contains anomalies.
This is common in environmental data like rainfall, where only a small percentage of years or months exhibit extreme or unusual patterns (e.g., droughts or extreme rainfall events).
Balancing Sensitivity:

A lower value (e.g., 0.01) would detect only the most extreme anomalies but may miss less obvious ones.
A higher value (e.g., 0.1) would classify more data points as anomalies, potentially including normal data as false positives.
Setting 0.05 strikes a balance, detecting meaningful outliers without being overly sensitive.
Domain Knowledge:

In climate studies, it's generally expected that anomalies (e.g., unusual rainfall patterns) occur infrequently. A value of 0.05 aligns with this expectation.
Adjusting this parameter should be based on expert knowledge or exploratory data analysis to validate the proportion of anomalies.
Algorithm Behavior:

Isolation Forest uses the contamination parameter to calculate a decision threshold internally. This ensures that approximately 5% of the data points are flagged as anomalies.
When to Adjust contamination?
Increase (e.g., 0.1 or higher):
If exploratory analysis shows a higher proportion of anomalies (e.g., frequent extreme rainfall events).
Decrease (e.g., 0.01 or lower):
If only the most extreme outliers are of interest, or the dataset is highly homogeneous.
Practical Implications in This Context:
Rainfall Data: Setting contamination=0.05 ensures that both extreme drought years and extreme rainfall years are captured effectively without overloading the result

s with false positives.

Determining the appropriate contamination value in IsolationForest depends on a combination of domain knowledge, exploratory data analysis, and experimentation. Here’s how you can systematically decide the best value:

1. Domain Knowledge
Use Prior Experience or Standards:
In fields like climatology, finance, or health, there are often known percentages of anomalies in similar datasets. For example:
Rainfall data: Anomalies (extreme drought or rainfall) might occur ~5% of the time.
Fraud detection: Fraudulent transactions are typically 1–2%.
Expert Input:
Consult domain experts or reference studies to estimate the proportion of anomalies expected in your dataset.
2. Analyze the Data
Perform exploratory data analysis (EDA) to understand the dataset's distribution and variability.

a. Visualize the Data:
Use box plots, histograms, or scatter plots to identify potential outliers visually.
Extreme values in the tails of the distribution may indicate anomalies.
b. Statistical Summary:
Compute statistics like the mean, standard deviation, and percentiles.
For instance, values beyond mean ± (k * std_dev) (e.g., k=1.5 or k=2) can help estimate the proportion of anomalies.

Experimental results.

Now, by applying k-means clustering, we will group years into three categories: Dry, Normal, and Wet, based on rainfall patterns

```python
from sklearn.cluster import KMeans  # Import KMeans for clustering data into distinct groups
from sklearn.preprocessing import StandardScaler  # Import StandardScaler for data normalization

# Prepare data for clustering

# Select relevant features for clustering
rainfall_features = rainfall_data[['Jan-Feb', 'Mar-May', 'Jun-Sep', 'Oct-Dec', 'ANNUAL']]
# Using seasonal and annual rainfall data to identify patterns and similarities between years

# Scale the features to standardize the data
scaler = StandardScaler()  # Initialize a StandardScaler to normalize data
scaled_features = scaler.fit_transform(rainfall_features)
# StandardScaler transforms the data to have a mean of 0 and standard deviation of 1.
# Why? Standardization ensures that all features contribute equally to the clustering.

# Perform K-Means clustering

# Initialize the KMeans algorithm with 3 clusters
kmeans = KMeans(n_clusters=3, random_state=42)
# n_clusters=3 assumes three distinct categories (e.g., Dry, Normal, Wet).
# random_state ensures reproducibility of the clustering results.

# Fit the KMeans model to the data and assign cluster labels
rainfall_data['Rainfall_Cluster'] = kmeans.fit_predict(scaled_features)
# fit_predict clusters the data and assigns a cluster label (0, 1, or 2) to each row.

# Map cluster labels to categories (e.g., Dry, Normal, Wet)

# Define a mapping for cluster labels to meaningful categories
cluster_labels = {0: 'Dry', 1: 'Normal', 2: 'Wet'}
# Why? Numerical labels are less interpretable, so mapping provides meaningful context.

# Create a new column with human-readable categories
rainfall_data['Rainfall_Category'] = rainfall_data['Rainfall_Cluster'].map(cluster_labels)
# map() converts the numerical cluster labels into their corresponding category names (e.g., 'Dry', 'Normal', 'Wet').
```

# Forecasting Future Rainfall
Finally,Ie will use the Prophet library to forecast annual rainfall for the next 20 years

```python
# Convert the 'YEAR' column into a datetime object
rainfall_data['DATE'] = pd.to_datetime(rainfall_data['YEAR'], format='%Y')
# Why? Prophet requires a datetime format for time-series data.

# Set 'DATE' as the index and extract the 'ANNUAL' column for time-series analysis
annual_rainfall_ts = rainfall_data.set_index('DATE')['ANNUAL']
# Why? Setting the datetime index organizes the data for forecasting.

# Install the Prophet library for time-series forecasting
# Use: pip install prophet
# Why? Prophet is an open-source library designed for accurate time-series forecasting.

from prophet import Prophet  # Import Prophet for creating forecasting models

# Prepare the data for Prophet
prophet_data = annual_rainfall_ts.reset_index()  # Reset the index to use 'DATE' as a column
prophet_data.columns = ['ds', 'y']
# Rename columns to 'ds' (datetime) and 'y' (value) as required by Prophet.

from prophet.plot import plot_plotly, plot_components_plotly  # Import plotting functions for visualizing forecasts

# Initialize a Prophet model
prophet_model = Prophet()
# Why? Prophet automatically handles seasonality, trends, and holidays, making it ideal for time-series data.

# Fit the Prophet model to the data
prophet_model.fit(prophet_data)
# Why? Fits the historical rainfall data to learn patterns and trends for forecasting.

# Create a future DataFrame for forecasting the next 20 years
future = prophet_model.make_future_dataframe(periods=20, freq='Y')
# periods=20: Forecast 20 years into the future.
# freq='Y': Use yearly frequency for the forecast.

# Predict future values based on the trained model
forecast = prophet_model.predict(future)
# Why? Generates predictions (with confidence intervals) for the future years based on the trained model.
```

```python
# Create an interactive plot of the forecast using Prophet's Plotly functionality
fig_forecast = plot_plotly(prophet_model, forecast)
# Why? `plot_plotly` generates an interactive visualization of the historical data, forecasted values,
# and confidence intervals for better analysis.

# Update the layout of the forecast figure for customization and clarity
fig_forecast.update_layout(
    title='Annual Rainfall Forecast Using Prophet',  # Add a descriptive title to the plot
    xaxis_title='Year',  # Label the x-axis as 'Year' to represent the timeline
    yaxis_title='Rainfall (mm)',  # Label the y-axis as 'Rainfall (mm)' for measurement units
    template='plotly_white',  # Use a clean white background for better readability
    height=500  # Set the height of the plot to 500 pixels for optimal viewing size
)

# Display the interactive forecast plot
fig_forecast.show()
# Why? Renders the chart, showing historical data, forecasted values, and uncertainty intervals
# to make the predictions interpretable and actionable.
```

The blue line represents the model’s forecast trend, while the shaded area indicates the confidence interval. The trend reveals a slight decline in annual rainfall over time, with notable year-to-year variability (black dots representing actual data points). The model captures the variability well but highlights that future rainfall may continue to slightly decrease, which emphasizes the need for adaptive strategies to manage potential water resource challenges.
